## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import pandas as pd

# Smart path detection for Google Colab vs Local execution
if os.path.exists("../../data/raw/content_refresh_anonymized.csv"):
    # Local relative path from work/notebooks/
    csv_path = "../../data/raw/content_refresh_anonymized.csv"
elif os.path.exists("data/raw/content_refresh_anonymized.csv"):
    # If running from the root directory
    csv_path = "data/raw/content_refresh_anonymized.csv"
else:
    # Fallback: Pull directly from your public GitHub repository raw link
    csv_path = "https://raw.githubusercontent.com/Bibek-Dhakal/ml-playground-for-applied-search-intelligence/main/data/raw/content_refresh_anonymized.csv"

# Load the starter dataset
df = pd.read_csv(csv_path)

# Filter for pages with actual visibility to avoid noise
df_visible = df[df['impressions_90d'] >= 100]

total_pages = len(df_visible)
avg_ctr = df_visible['ctr'].mean()
stale_pages = len(df_visible[df_visible['days_since_last_update'] >= 180])

print(f"--- Quick Data Reality Check ---")
print(f"Successfully loaded dataset from: {csv_path}")
print(f"1. There are {total_pages} pages in the starter dataset with enough visibility (>= 100 impressions) to safely analyze.")
print(f"2. The baseline average CTR across these visible pages is {avg_ctr:.3f}.")
print(f"3. Out of those visible pages, {stale_pages} have not been updated in over 6 months, representing an immediate potential target pool.")
```
